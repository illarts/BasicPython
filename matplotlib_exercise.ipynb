## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('data/hist_data.csv')

# ラベルとかいろいろ
plt.title("aデータとｂデータの温度に関する度数分布表", fontsize=15)
plt.xlabel("温度(°C)", fontsize=10)
plt.ylabel("度数", fontsize=10) 

# ヒストグラムを作成する
plt.hist(
    df['aデータ'],
    bins=30,
    alpha=0.5,
    color= 'orange',
    label="aデータ"
    )

plt.hist(
    df['bデータ'],
    bins=30,
    alpha=0.5,
    color= 'c',
    label="bデータ"
    )

plt.legend(loc="upper right", fontsize=13)

# https://ai-inter1.com/python-histogram/
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv('data/scatter_data.csv')
# length (cm),width (cm),class

# クラスごとにデータを分割する
groups = df.groupby('class')

label2color = ['r', 'b', 'g']

for cls, group, in groups:
    plt.scatter(
        group['length (cm)'],
        group['width (cm)'],
        label = "class" + str(cls),
        c = label2color[cls] # クラスごとに散布図をプロット
        )

# ラベルとかいろいろ
plt.title("クラス０～２のlengthとwidthに関する散布図", fontsize=15)
plt.xlabel("length(cm)", fontsize=10)
plt.ylabel("width(cm)", fontsize=10) 

# 凡例を表示する
plt.legend(loc="upper right", fontsize=13)

# グリッドを表示する
plt.grid(True)

# グラフを表示する
plt.show()

# https://qiita.com/y629/items/afa82bd4145f940cb28b
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv('data/line_data.csv')
# 期間,件数,費用

plt.plot(
    df["期間"],
    df["件数"],
    
    marker='o',
    linestyle='-'
    )

plt.xticks(rotation = 30)

# ラベルとかいろいろ
plt.title("期間ごとの件数の推移", fontsize=15)
plt.xlabel("期間", fontsize=10)
plt.ylabel("件数", fontsize=10) 

# グリッドを表示する
plt.grid(True)

# グラフを表示する
plt.show()

##################

plt.bar(
    df["期間"],
    df["費用"],
    )

plt.xticks(rotation = 30)

# ラベルとかいろいろ
plt.title("期間ごとの件数の推移", fontsize=15)
plt.xlabel("期間", fontsize=10)
plt.ylabel("費用（円）", fontsize=10) 

# グリッドを表示する
plt.grid(True)

# グラフを表示する
plt.show()
```
