### 問3. データの可視化（Extra）

3.1. x 軸を petal length (cm)、y 軸を petal width (cm)にして散布図を作成しましょう。 ただし、目的変数に応じて点の色と形を変更してください。

```python
import matplotlib.pyplot as plt

# Irisデータセットの読み込み
iris = load_iris()
X = iris.data[:, 2:4]  # petal length (cm) と petal width (cm)
y = iris.target

# 目的変数に応じた色と形のリストを作成
colors = ['red', 'green', 'blue']

# 散布図をプロット
for i in range(len(colors)):
    plt.scatter(X[y == i, 0], X[y == i, 1], c=colors[i], label=iris.target_names[i])

# 軸ラベルと凡例の追加
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()

# グラフを表示
plt.show()
```

### 問4. データの分割

4.1. iris データを 4:1 に分割して、80% を学習データとして 20% をテストデータとしてください。

ただし、分割した後の変数名は以下の通りにしてください。

| 変数名  |      内容      |
| :-----: | :------------: |
| x_train | 学習用の説明変数 |
| x_test | テスト用の説明変数 |
| y_train | 学習用の目的変数 |
| y_test | テスト用の目的変数 |

学習データでモデルの学習を行い、テストデータでモデルの評価を行います。


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Irisデータセットを読み込む
iris = load_iris()
x = iris.data
y = iris.target

# 4:1の比率でデータセットを学習データとテストデータに分割する
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

4.2. 本当に4:1に分割できているか確認しましょう。

```python

# 分割されたデータセットの形状を表示して、分割が正しく行われたことを確認する
print("x_train:", x_train.shape)
print("x_test:", x_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

### 問5. 学習

5.1. RandamForest の学習モデルのインスタンスを生成しましょう。

```python
from sklearn.ensemble import RandomForestClassifier

# RandomForestClassifierのインスタンスを生成する
rf_classifier = RandomForestClassifier()

# モデルの学習を行うためにfitメソッドを使用し、学習データを渡す
rf_classifier.fit(x_train, y_train)
```

5.2. `x_train` と `y_train` を用いて、モデルを学習させましょう。

```python
# RandomForestClassifierのインスタンスを生成する
rf_classifier = RandomForestClassifier()

# モデルの学習を行うためにfitメソッドを使用し、学習データを渡す
rf_classifier.fit(x_train, y_train)
```

### 問6. 推論


6.1. 学習したモデルに `x_test` を入力して予測をしましょう。

```python
# テストデータを使って予測を行う
y_pred = rf_classifier.predict(x_test)
```

6.2. 予測結果と `y_test` に対して、`accuracy` を計算することでモデルを評価しましょう。

```python
from sklearn.metrics import accuracy_score

# 予測結果と実際のラベルを比較し、精度を計算する
accuracy = accuracy_score(y_test, y_pred)

# 精度を出力する
print("Accuracy:", accuracy)
```

実は、iris データセットはかなり簡単な分布をしているので、正解率はほぼ 100% になったと思います。
